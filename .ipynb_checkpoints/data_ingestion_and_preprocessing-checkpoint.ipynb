# Data Ingestion and Preprocessing

Reads all dataset, preprocess them and fit into the schema for the analysis_of_bandit_algorithms jupyter notebook. Outputs preprocessed datasets into the data folder.

```python
import pandas as pd
```

```python
# All required directories / files
data_dir = './data/'
movie_review_dataset = 'top-n-movies_user-ratings.csv'
movie_review_output_path = './data/preprocessed_movie_ratings.csv'
```

## Movie Review Dataset

```python
# Movie Reviews of Top 10 Movies
movie_reviews_df = pd.read_csv(data_dir + movie_review_dataset)
# movie_reviews_df.drop(movie_reviews_df.columns[0], inplace=True, axis=1)
movie_reviews_df
```

```python
# Top 10 movies
movie_reviews_df['movie_title'].unique()
```

```python
# Generate the reward column
threshold_val_for_reward = 5
movie_reviews_df['reward'] = movie_reviews_df['rating'].apply(lambda x: 0 if x<threshold_val_for_reward else 1) 

# Generate rewards probabilities for each movie
movie_reviews_output_df = pd.DataFrame(movie_reviews_df.groupby('movie_title').apply(lambda x: sum(x['reward'])/len(x))).reset_index()
movie_reviews_output_df = movie_reviews_output_df.rename({0: 'reward_probabilities'}, axis=1)
movie_reviews_output_df
```

```python
# Outputs dataframe to csv
movie_reviews_output_df.to_csv(movie_review_output_path, index=False)
```
